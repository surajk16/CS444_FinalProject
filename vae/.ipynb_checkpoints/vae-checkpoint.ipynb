```python
import torch
from torchvision import datasets, transforms
from torch import optim

import matplotlib.pyplot as plt

import time

from vae import VAE

%load_ext autoreload
%autoreload 2
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
batch_size = 32
data_dir = "../data/all-dogs/"

random_transforms = [transforms.ColorJitter(), transforms.RandomRotation(degrees=20)]
transform = transforms.Compose([
    transforms.Resize(64),
    transforms.CenterCrop(64),
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.RandomApply(random_transforms, p=0.2),
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))]
)

train_data = datasets.ImageFolder(data_dir, transform=transform)
train_loader = torch.utils.data.DataLoader(train_data, shuffle=True, batch_size=batch_size)
                                           
imgs, label = next(iter(train_loader))
imgs = imgs.numpy().transpose(0, 2, 3, 1)
```

```python
fig = plt.figure(figsize=(25, 16))
for i, img in enumerate(imgs):
    ax = fig.add_subplot(4, 8, i + 1, xticks=[], yticks=[])
    
    plt.imshow((img+1)/2)
```

```python
lr = 0.001
epochs = 25
latent_dim = 32

model = VAE(latent_dim, batch_size=batch_size).to(device)
optimizer = optim.Adam(model.parameters(), lr=lr)
```

```python
plt.imshow((imgs[0]+1)/2)
plt.show()
```

```python
for epoch in range(1, epochs+1):
    start_time = time.time()
    model.train()
    print('Starting epoch: {}'.format(epoch))
    
    mse_loss_total = 0
    kld_loss_total = 0
    
    for batch_idx, (data, labels) in enumerate(train_loader):
        data = data.to(device)
        optimizer.zero_grad()

        recon_batch, mu, logvar = model(data)
        mse_loss, kld_loss = model.loss_function(recon_batch, data, mu, logvar)
        
        mse_loss_total += mse_loss
        kld_loss_total += kld_loss
        
        loss = mse_loss + kld_loss

        loss.backward()
        optimizer.step()
    
    print("Loss: Total: {}, MSE: {}, KLD: {}".format(loss, mse_loss_total, kld_loss_total))
    print('Time for epoch {}: {}'.format(epoch, time.time()-start_time))
        
    model.eval()
    recon_img, _, _ = model(torch.tensor(imgs[:1].transpose(0, 3, 1, 2)).to(device))
    img = recon_img.view(3, 64, 64).detach().cpu().numpy().transpose(1, 2, 0)
    
    plt.imshow((img+1.)/2.)
    plt.show()
```
